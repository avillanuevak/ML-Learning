## Gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt
x= np.array([1, 2, 3, 4, 5])
y= np.array([2, 4, 3, 7, 9])
```

```python
def gradient_descent(x, y, learning_rate=0.01, iterations=1000):
    m = 0  # slope
    b = 0  # intercept
    n = len(y)  # number of data points

    for _ in range(iterations):
        y_pred = m * x + b  # predicted values
        error = y_pred - y  # error
        cost = (1/n) * np.sum(error ** 2)  # cost function

        # gradients
        m_gradient = (2/n) * np.dot(x, error)
        b_gradient = (2/n) * np.sum(error)

        # update parameters
        m -= learning_rate * m_gradient
        b -= learning_rate * b_gradient

    return m, b,cost
```

```python
gradient_descent_result = gradient_descent(x, y)
print(f"Slope: {gradient_descent_result[0]}, Intercept: {gradient_descent_result[1]}, Cost: {gradient_descent_result[2]}")
# Plotting the results
plt.scatter(x, y, color='blue', label='Data points')
m, b,_ = gradient_descent_result
plt.plot(x, m * x + b, color='red', label='Fitted line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Linear Regression')
plt.legend()
plt.show()
```
